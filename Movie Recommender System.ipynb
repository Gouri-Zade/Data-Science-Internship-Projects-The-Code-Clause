# Column Extraction To Transformation into One Variable
* The Overview of Column Extraction : I have extracted the relevant data from "generes, keyword, cast and crew" features to transform them into "tags feature" for best recommendation system.

## Column Extraction

### 1) Column Extraction : Genere Feature

```python
# example : genre info of 1st movie
movies_data.genres.iloc[0]
```

* Extracting the value of all 'name' key from the list of dictionaries. After extracting the genre feature look like this:

      ['Action', "Adventure", 'Fantasy' , 'Science Fiction']

```python
import ast # to convert the string into list of dictionaries

# creat function to extract the value of "name" key
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies_data['genres'] = movies_data['genres'].apply(convert)
```

### 2) Column Extraction : Keyword Feature

```python
# example : keyword info of 1st movie
movies_data['keywords'].iloc[0]
```

* Extracting the value of all 'name' key from the list of dictionaries.

```python
movies_data['keywords'] = movies_data['keywords'].apply(convert)
```

### 3) Column Extraction : Cast Feature

```python
# # example : cast info of 1st movie
movies_data['cast'].iloc[0]
```

* Extracting the top 5 cast from cast feature.

```python
# creat function to extract the value of "name" key
def top5_Cast(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 5:
      l.append(i['name'])
      counter = counter + 1
    else:
      break
  return l
```

```python
movies_data['cast'] = movies_data['cast'].apply(top5_Cast)
```

### 4) Column Extraction : Crew Feature

```python
# example : crew info of 1st movie
movies_data['crew'].iloc[0]
```

* Extracting the name from crew feature where job of the crew is 'Director'.

```python
# creat function to extract the name of director
def director(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
```

```python
movies_data['crew'] = movies_data['crew'].apply(director)
```

## Feature Transformation

### 1) Overview Feature
* Spitting the feature to convert it into one feature

```python
movies_data['overview'] = movies_data['overview'].apply(lambda x: x.split())
```

```python
movies_data.head(3)
```

### Replacing the Spaces

```python
movies_data['genres'] = movies_data['genres'].apply(lambda x:[i.replace(" ", '') for i in x]) # genres feature
movies_data['keywords'] = movies_data['keywords'].apply(lambda x:[i.replace(" ", '') for i in x]) # keywords feature
movies_data['cast'] = movies_data['cast'].apply(lambda x:[i.replace(" ", '') for i in x]) # cast feature
movies_data['crew'] = movies_data['crew'].apply(lambda x:[i.replace(" ", '') for i in x]) # crew feature
```

* All the spaces have been removed to make recommendation better.

```python
movies_data.head(3)
```

### Concating all the extracting feature to transform in one variable

```python
movies_data['tags'] = movies_data['overview'] + movies_data['genres'] + movies_data['keywords'] + movies_data['cast'] + movies_data['crew']
```

* All features have transformed in 'tags' feature.

```python
movies_data.head()
```

* Taking the useful feature for analysis

```python
new_movies_data = movies_data[['movie_id', 'title', 'tags']]
new_movies_data.head(3)
```

### Converting the tags column from list into one string

```python
new_movies_data['tags'] = new_movies_data['tags'].apply(lambda x: " ".join(x)) # to convert into string
```

```python
new_movies_data['tags'] = new_movies_data['tags'].apply(lambda x: x.lower()) # convering to lower case
```

```python
#  1st movie tag
new_movies_data['tags'].iloc[0]
```

# Stemming
*  Stemming is used to normalize or standardize words.
* Stemming refers to the process of reducing words to their base or root form, known as the stem.
* stemming helps in consolidating and collapsing similar words, reducing the vocabulary size and improving computational efficiency.

```python
import nltk
from nltk.stem.porter import PorterStemmer # PorterStemmer technique is used for stemming
ps = PorterStemmer()
```

```python
# creat function to stem the texts
def stem(text):
  tags = []
  for i in text.split(): # split the string and convert to list
    tags.append(ps.stem(i))

  return " ".join(tags)
```

```python
# apply the stemming
new_movies_data['tags'] = new_movies_data['tags'].apply(stem)
```

# Vectorization
* Vectorization refers to the process of representing textual data such as words, sentences or document as numerical vectors.
* CountVecterizer technique is used in this recommendation

```python
# import CountVertorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000, stop_words='english')
```

* Top 5000 most occuring tag words are taken to make best recommendation

```python
# converting vectors into array format
vectors = cv.fit_transform(new_movies_data['tags']).toarray()
vectors
```

```python
# 1st movie tags
vectors[0]
```

```python
cv.get_feature_names_out()
```

### Calculating the similarity between the tag words to recommend the movie name
* Here Cosine Similarity is used to find the most similar tag words

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity.shape
```

# Movie Recommendation System

```python
# Sorting the similarty movie with 5 recommendation
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])[1:6]
```

```python
# Create function to recommond movie name
def recommend(movie):
  movies_index = new_movies_data[new_movies_data['title'] == movie].index[0]
  distances = similarity[movies_index]

  # Sorting the similarty movie with 5 recommendation
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]

  for i in movies_list:
    print(new_movies_data.iloc[i[0]].title) # print the movie title
```

# Recommend Top five movies name based on the given name

```python
# example 1
recommend('Avatar')
```

```python
# example 2
recommend('The Smurfs 2')
```
